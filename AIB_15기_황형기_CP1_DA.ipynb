```python
import matplotlib.pyplot as plt
import seaborn as sns

df_2000 = pd.read_csv(r'C:\Users\paulhwang\cp1\1인가구비율_2000년.csv', encoding='cp949')


df_2000 = df_2000.astype({'2000':'float', '2000.1':'int', '2000.2':'int', '2005':'float', '2005.1':'int', 
'2005.2':'int', '2010':'float', '2010.1':'int', '2010.2':'int'})

trend = df_2000[['2000', '2005', '2010', '2015', '2016', '2017', '2018', '2019', '2020', '2021']][:1]
```

```python
trend.T.reset_index()
```

```python
# 1인가구 증가량 추세선

plt.figure(figsize=(10,5))
sns.lineplot(data=trend.T, markers=True, dashes=False)
plt.grid(True)

plt.show()
```

```python
ratio_2021 = df_2000[['2021.1', '2021.2']][:1].T
ratio_2021 / ratio_2021.sum()
```

```python
# 2021년 1인가구와 다인가구 비율 비교
x_label = ['a single-person household', 'multi-person household']

fig = plt.figure(figsize=(8,8))
fig.set_facecolor('white')
ax = fig.add_subplot()
pie = ax.pie(
    ratio_2021[0], 
    labels = x_label, 
    autopct = '%.1f%%', 
    explode = [0, 0.2],
    colors = ['lightskyblue', 'lightcoral'],
    shadow=True)
plt.legend()
plt.show()
```
