```python
def copmpanion_type_per(df):
    df_family = round(df[df['A7B_1']=='가족(부모, 배우자, 자녀, 형제)'].shape[0] / df.shape[0], 3)
    df_cousin = round(df[df['A7B_2']=='친척'].shape[0] / df.shape[0], 3)
    df_friends = round(df[df['A7B_3']=='친구/연인'].shape[0] / df.shape[0], 3)
    df_colleague = round(df[df['A7B_4']=='(직장)동료'].shape[0] / df.shape[0], 3)
    df_gathering = round(df[df['A7B_5']=='친목 단체/모임(동호회, 종교단체 등)'].shape[0] / df.shape[0], 3)
    df_school = round(df[df['A7B_6']=='학교 단체(소풍, 수학여행 등)'].shape[0] / df.shape[0], 3)
    df_other = round(df[df['A7B_7']=='기타'].shape[0] / df.shape[0], 3)

    return df_family, df_cousin, df_friends, df_colleague, df_gathering, df_school, df_other
```

```python
def divide_age(df):
    under_30 = df[(df['BAGE']=='  15~19세') | (df['BAGE']=='  20대') | (df['BAGE']=='  30대')]
    from_40_to_50 = df[(df['BAGE']=='  40대') | (df['BAGE']=='  50대')]
    over_60 = df[(df['BAGE']=='  60대') | (df['BAGE']=='  70세이상')]

    return under_30, from_40_to_50, over_60

oph_under_30_2019, oph_4_50_2019, oph_over_60_2019 = divide_age(oph_trav_exp_2019)
oph_under_30_2020, oph_4_50_2020, oph_over_60_2020 = divide_age(oph_trav_exp_2020)
oph_under_30_2021, oph_4_50_2021, oph_over_60_2021 = divide_age(oph_trav_exp_2021)

mph_under_30_2019, mph_4_50_2019, mph_over_60_2019 = divide_age(mph_trav_exp_2019)
mph_under_30_2020, mph_4_50_2020, mph_over_60_2020 = divide_age(mph_trav_exp_2020)
mph_under_30_2021, mph_4_50_2021, mph_over_60_2021 = divide_age(mph_trav_exp_2021)
```

```python
def advent_type(df):
    nature = round((len(df['A5_1']) - df['A5_1'].isnull().sum()) / df.shape[0],3)
    food = round((len(df['A5_2']) - df['A5_2'].isnull().sum()) / df.shape[0],3)
    leisure = round((len(df['A5_3']) - df['A5_3'].isnull().sum()) / df.shape[0],3)
    history = round((len(df['A5_4']) - df['A5_4'].isnull().sum()) / df.shape[0],3)
    amusement = round((len(df['A5_5']) - df['A5_5'].isnull().sum()) / df.shape[0],3)
    rest = round((len(df['A5_6']) - df['A5_6'].isnull().sum()) / df.shape[0],3)
    culture = round((len(df['A5_9']) - df['A5_9'].isnull().sum()) / df.shape[0],3)
    fam_friend = round((len(df['A5_18']) - df['A5_18'].isnull().sum()) / df.shape[0],3)

    return nature, food, leisure, history, amusement, rest, culture, fam_friend
```

```python
def divide_income(df):
    under_100 = df[(df['DQ6A']=='소득 없음') | (df['DQ6A']=='월평균 100만원 미만')]
    from_100_to_300 = df[(df['DQ6A']=='월평균 100만원 ~ 200만원 미만') | (df['DQ6A']=='월평균 200만원 ~ 300만원 미만')]
    from_300_to_500 = df[(df['DQ6A']=='월평균 300만원 ~ 400만원 미만') | (df['DQ6A']=='월평균 400만원 ~ 500만원 미만')]
    over_500 = df[(df['DQ6A']=='월평균 500만원 ~ 600만원 미만') | (df['DQ6A']=='월평균 600만원 ~ 700만원 미만') | (df['DQ6A']=='월평균 700만원 ~ 800만원 미만') | (df['DQ6A']=='월평균 800만원 ~ 900만원 미만') | (df['DQ6A']=='월평균 1,000만원 이상')]

    return under_100, from_100_to_300, from_300_to_500, over_500
```

```python
def rename_col(df):
    df['BSEX'] = df['BSEX'].replace({'  남자':1, '  여자':2})
    df['BAGE'] = df['BAGE'].replace({'  15~19세':1, '  20대':1, '  30대':1, '  40대':2, '  50대':2, '  60대':3, '  70세이상':3})
    df['BEDU'] = df['BEDU'].replace({'  초졸 이하':1, '  중학교':2,'  고등학교':3, '  대학교이상':4 })
    df['BMAR'] = df['BMAR'].replace({'  미혼':1, '  배우자 있음':2,'  사별':3, '  이혼':4 , '  기타':5})
    df['BJOB'] = df['BJOB'].replace({'  임금봉급근로자':1, '  고용원있는사업주':2,'  고용원없는자영자':3, '  무급가족 종사자':4 , '  전업주부':5, '  학생':6, '  기타':7})
    df['DQ6B'] = df['DQ6B'].replace(
        {
            '소득없음':1, '월평균 100만원 미만':1, '월평균 100만원 ~ 200만원 미만':1,'월평균 200만원 ~ 300만원 미만':1, 
            '월평균 400만원 ~ 500만원 미만':2 , '월평균 300만원 ~ 400만원 미만':2, 
            '월평균 500만원 ~ 600만원 미만':3, '월평균 600만원 ~ 700만원 미만':3,'월평균 600만원 ~ 700만원 미만':3, '월평균 700만원 ~ 800만원 미만':3, '월평균 800만원 ~ 900만원 미만':3, '월평균 1,000만원 이상':3
            }
            )
    df['국내_연간여행경험_국내전체'] = df['국내_연간여행경험_국내전체'].replace({100.0:1, 0.0:2, '경험':1, '비경험':2})
    df.rename(columns={'국내_연간여행경험_국내전체':'LABEL'}, inplace=True)
    df['LABEL'] = df['LABEL'].astype(int)

    return df
```
